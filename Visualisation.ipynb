```python
file_path = "C:\\Users\\PALLAVI\\OneDrive\\Documents\\CleanedDataset.csv"
df = pd.read_csv(file_path)
df.columns = df.columns.str.strip()

# If multiple entries per subdivision, group and take average
avg_levels = df.groupby('SUBDIVISION')['Level'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x="SUBDIVISION", y="Level", data=avg_levels, marker='o', color='purple')

plt.xticks(rotation=45, ha="right")  
plt.xlabel("SUBDIVISION")
plt.ylabel("Average Level")
plt.title("Average Level across SUBDIVISIONS (Line Graph)")
plt.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("C:\\Users\\PALLAVI\\OneDrive\\Documents\\CleanedDataset.csv")
data.columns = data.columns.str.strip()

months = ["JUN", "JUL", "AUG", "SEP"]
rainfall = data[months].sum()  # Summing up rainfall across all reservoirs

plt.pie(rainfall, labels=months, autopct='%1.1f%%', startangle=90)
plt.title("Rain Data Set")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[['JUN', 'JUL', 'AUG', 'SEP', 'June-September', 
                'Temperature Min', 'Temperature Max', 
                'pH Min', 'pH Max',
                'Full_reservoir_level', 'Live_capacity_FRL', 
                'Storage', 'Level']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='SUBDIVISION', y='Level', data=df)
plt.xticks(rotation=90)
plt.title("Level by SUBDIVISION")
plt.show()
```

```python
for col in ['JUN', 'JUL', 'AUG', 'SEP', 'June-September', 
            'Temperature Min', 'Temperature Max', 
            'pH Min', 'pH Max', 'Full_reservoir_level', 
            'Live_capacity_FRL', 'Storage', 'Level']:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
plt.figure(figsize=(12,6))
avg_levels = df.groupby('SUBDIVISION')['Level'].mean().sort_values()
sns.barplot(x=avg_levels.index, y=avg_levels.values, palette='viridis')
plt.xticks(rotation=90)
plt.title("Average Level by SUBDIVISION")
plt.ylabel("Average Level")
plt.xlabel("SUBDIVISION")
plt.show()
```

```python
monthly_cols = ['JUN', 'JUL', 'AUG', 'SEP']
for col in monthly_cols:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df[col], y=df['Level'])
    plt.title(f"{col} vs Level")
    plt.xlabel(col)
    plt.ylabel("Level")
    plt.show()
```
